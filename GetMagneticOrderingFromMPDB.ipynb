```python
from pymatgen import MPRester, Composition
from pymatgen.electronic_structure.plotter import BSPlotter
import pymatgen.analysis.find_dimension
import pymatgen.io.cif as pcif
import pandas as pd
import pymatgen.analysis.find_dimension as pfd
from tqdm import tqdm
import pymatgen.analysis.magnetism.analyzer as pama
```

```python
df_1D = pd.read_excel('Cheno_1D_2D_withMag_v2.xlsx',sheet_name='1D')
df_2D = pd.read_excel('Cheno_1D_2D_withMag_v2.xlsx',sheet_name='2D')
df_q1D = pd.read_excel('Cheno_1D_2D_withMag_v2.xlsx',sheet_name='quasi-1D')
```

```python
ordering_1D = []
ordering_2D = []
ordering_q1D = []
mpr = MPRester("Frv0akZ1InOToUmL")##API key should be inside the "", URL for API key: https://materialsproject.org/dashboard
```

```python
for i in tqdm(range(len(df_1D['mpid'].values))):
    mpid = df_1D['mpid'][i]
    try:
        structure = mpr.get_structure_by_material_id(mpid)
        #structures.append(structure)
        #magmoms.append(structure.site_properties['magmom'])
        ordering_1D.append(pama.CollinearMagneticStructureAnalyzer(structure).ordering)
    except:
        ordering_1D.append(None)
        #structures.append(None)
```

```python
for i in tqdm(range(len(df_2D['mpid'].values))):
    mpid = df_2D['mpid'][i]
    try:
        structure = mpr.get_structure_by_material_id(mpid)
        #structures.append(structure)
        #magmoms.append(structure.site_properties['magmom'])
        ordering_2D.append(pama.CollinearMagneticStructureAnalyzer(structure).ordering)
    except:
        ordering_2D.append(None)
        #structures.append(None)
```

```python
for i in tqdm(range(len(df_q1D['mpid'].values))):
    mpid = df_q1D['mpid'][i]
    try:
        structure = mpr.get_structure_by_material_id(mpid)
        #structures.append(structure)
        #magmoms.append(structure.site_properties['magmom'])
        ordering_q1D.append(pama.CollinearMagneticStructureAnalyzer(structure).ordering)
    except:
        ordering_q1D.append(None)
        #structures.append(None)
```

```python
def ordering_value(ordering):
    ordering_val = []
    for order in ordering:
        try:
            ordering_val.append(order.value)
        except:
            ordering_val.append(None)
    return ordering_val
```

```python
ordering_val_1D = ordering_value(ordering_1D)
ordering_val_2D = ordering_value(ordering_2D)
ordering_val_q1D = ordering_value(ordering_q1D)
```

```python
df_ordering_1D = pd.DataFrame({'mpid':df_1D['mpid'],'pretty_formula':df_1D['pretty_formula_MP'],'band_gap':df_1D['band_gap_MP'],'total_magnetization':df_1D['total_magnetization_MP'],'ordering':ordering_val_1D})
```

```python
df_ordering_2D = pd.DataFrame({'mpid':df_2D['mpid'],'pretty_formula':df_2D['pretty_formula_MP'],'band_gap':df_2D['band_gap_MP'],'total_magnetization':df_2D['total_magnetization_MP'],'ordering':ordering_val_2D})
df_ordering_q1D = pd.DataFrame({'mpid':df_q1D['mpid'],'pretty_formula':df_q1D['pretty_formula_MP'],'band_gap':df_q1D['band_gap_MP'],'total_magnetization':df_q1D['total_magnetization_MP'],'ordering':ordering_val_q1D})
```
